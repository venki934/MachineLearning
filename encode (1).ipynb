```python
from sklearn.preprocessing import LabelEncoder
```

```python
data=["small","medium","large","medium"]
le=LabelEncoder()
encoded=le.fit_transform(data)
print(encoded)
```

```python
data=np.array([["red"],["blue"],["green"]])
encoder=OneHotEncoder()
encoded=encoder.fit_transform(data).toarray()
print(encoded)
```
